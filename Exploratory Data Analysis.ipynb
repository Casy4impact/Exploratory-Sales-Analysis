# Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) is the process of analysing and sumarising data in order to gain insights and understanding of the underlying patterns and relationships. The main aim of EDA is to identify and explore the main charactersitics and patterns of the data, and to identify any anomalies or outliers that may impact subsequent analysis. 

EDA typically involves a number of steps, including

1. Data cleaning: Data cleaning involves removing or correcting any errors or inconsistencies in the data, such as missing values or incorrect values
2. Data visualisation: Data visualisation techniques are then used to graphically represent the data and identify any trends or patterns
3. Statistical analysis: Statistical analysis is used to identify any relationships between variables and to test hypotheses about the data. This may involve calculating summary statistics such as mean and standard deviation, and performing test such as correlation analysis and hypothesis testing.

# Dealing with missing data -

- 1.	MCAR (Missing completely at random): These are values that are randomly missing and do not depend on any other values
- 2.	MAR (Missing at random): These values are dependent on some additional features
- 3.	MNAR (Missing not at random): There is a reason behind these values are missing

There are several methods for imputing missing data, including the measures of Central Tendencies, regression imputation, and multiple imputation. Measures of Central Tendency involves replacing missing values with either the Mean, Median and Mode of the variable, while regression imputation involves using other variables in the dataset to predict missing values

```python
# Convert the data column into a pandas datetime object
df["Date"]=pd.to_datetime(df["Date"])

# Extract the Year, Month, Quarter and week
df['year']= df['Date'].dt.year
df['month']= df['Date'].dt.month
df['month_name']= df['Date'].dt.month_name()
df['quarter']= df['Date'].dt.quarter
# df['week']= df['Date'].dt.week
# Series.dt.isocalendar().week # NameError: name 'Series' is not defined
df.head(2)
```

```python
# Group Customer Age
def age_group(x):
    if x <= 25:
        return "<=25 Young Adult"
    elif x <=40:
        return "26-40 Adult"
    elif x <=50:
        return "41-50 old Adult"
    else:
        return ">= 51 Elder"

# Applying function to the data
df["age_group"] = df["Customer_Age"].apply(age_group)
df.tail(2)
```

```python
# Cost, Revenue and Profit Calculations

df["cost"] = df["Quantity"]*df["Unit_Cost"]
df["revenue"] = df["Quantity"]*df["Unit_Price"]
df["profit"] = df["revenue"]-df["cost"]

df.head(10)
```

```python
# Profit/Loss Grouping
'''
def porl(x):
    if x >=0:
        return "Profit"
    else:
        return "Loss"

df["profit_label"] = df["profit"].apply(porl)
df.head(2)
'''
```

```python
# Profit/Loss Grouping

# Define a function to calculate profit_label
def profit_loss(profit):
    if profit > 0:
        return "Profit"
    elif profit < 0:
        return "Loss"
    else:
        return "No Profit"

# Apply the function to create the "profit_label" column
df["profit_label"] = df["profit"].apply(profit_loss)
df.head(3)
```
